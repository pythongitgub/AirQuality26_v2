```

import requests
import pandas as pd

url = "https://uk-air.defra.gov.uk/sos-ukair/api/v1/stations"
stations = requests.get(url).json()

records = []
for s in stations.get("data", []):
    records.append({
        "station": s.get("name"),
        "lat": s.get("latitude"),
        "lon": s.get("longitude")
    })

df = pd.DataFrame(records)
df.to_csv("outputs/air_monitor_registry.csv", index=False)
```
